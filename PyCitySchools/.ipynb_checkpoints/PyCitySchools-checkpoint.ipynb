School name

School type

Total students

Per student budget

% passing math (the percentage of students who passed math)

```python


school_names = joined_schol_data.set_index("school_name").groupby(['school_name'])

school_types = school_data.set_index("school_name")["type"]

student_per_school = joined_schol_data.sort_values('school_name').set_index("school_name")["size"].drop_duplicates()

school_budget_column = joined_schol_data.set_index("school_name")['budget'].drop_duplicates()

school_budget_per_student = school_budget_column/student_per_school

avg_math_score = joined_schol_data.groupby(["school_name"])['math_score'].mean()

avg_reading_score = joined_schol_data.groupby(["school_name"])['reading_score'].mean()

math_pass_pct = joined_schol_data[joined_schol_data['math_score'] >= 70].groupby(["school_name"])["Student ID"].count()*100/student_per_school

reading_pass_pct = joined_schol_data[joined_schol_data['reading_score'] >= 70].groupby(["school_name"])["Student ID"].count()*100/student_per_school

math_read_pass_pct = joined_schol_data[(joined_schol_data['reading_score'] >= 70) & (joined_schol_data['math_score'] >= 70)].groupby(["school_name"])["Student ID"].count()*100/student_per_school

summary_school = pd.DataFrame({
    "School types": school_types,
    "Total Students": student_per_school,
    "Total school budget": school_budget_column,
    "Per Student Budget" : school_budget_per_student,
    "Average Math Score": avg_math_score,
    "Average Reading Score": avg_reading_score,
    "% Passing Math": math_pass_pct,
    "% Passing Reading": reading_pass_pct,
    "% Overall Passing": math_read_pass_pct
})

summary_school["Total school budget"] = summary_school.apply(lambda x: "${:,.2f}".format(x['Total school budget']), axis=1)
summary_school["Per Student Budget"] = summary_school.apply(lambda x: "${:,.2f}".format(x['Per Student Budget']), axis=1)
summary_school
```

Scores by School Spending
Create a table that breaks down school performance based on average spending ranges (per student). Use your judgment to create four bins with reasonable cutoff values to group school spending. Include the following metrics in the table:

Average math score

Average reading score

% passing math (the percentage of students who passed math)

% passing reading (the percentage of students who passed reading)

% overall passing (the percentage of students who passed math AND reading)

```python
# Create a table that breaks down school performances based on average Spending Ranges (Per Student). 
# Use 4 reasonable bins to group school spending. 
# Include in the table each of the following:
# Average Math Score
# Average Reading Score
# % Passing Math
# % Passing Reading
# Overall Passing Rate (Average of the above two)

# 

bins = [0, 585, 630, 645, 680]
names = ['<$585','$585-630', '$630-645','$645-680']

joined_schol_data['binned'] = pd.cut(
    joined_schol_data["budget"]/joined_schol_data["size"], bins, labels=names)

grouped_spending = joined_schol_data.groupby("binned")

avg_math_score = grouped_spending['math_score'].mean()

avg_reading_score = grouped_spending['reading_score'].mean()

math_pass_pct = joined_schol_data[joined_schol_data['math_score'] >= 70].groupby("binned")["Student ID"].count()*100/grouped_spending["Student ID"].count()

reading_pass_pct = joined_schol_data[joined_schol_data['reading_score'] >= 70].groupby("binned")["Student ID"].count()*100/grouped_spending["Student ID"].count()

math_read_pass_pct = joined_schol_data[(joined_schol_data['reading_score'] >= 70) & (joined_schol_data['math_score'] >= 70)].groupby("binned")["Student ID"].count()*100/grouped_spending["Student ID"].count()

scores_by_school_spending = pd.DataFrame({
    "Average Math Score": avg_math_score.round(2),
    "Average Reading Score": avg_reading_score.round(2),
    "% Passing Math": math_pass_pct.round(2),
    "% Passing Reading": reading_pass_pct.round(2),
    "% Overall Passing": math_read_pass_pct.round(2)
})

scores_by_school_spending
```

Scores by School Size
Create a table that breaks down school performance based on school size (small, medium, or large)

```python
# Scores by School Size
# Perform the same operations as above, based on school size.
brackets = [0, 1000, 2000, 5000]
school_size_label = ['Small (<1000)','Medium (1000-2000)', 'Large (2000-5000)']

joined_schol_data['binned'] = pd.cut(
    joined_schol_data["size"], brackets, labels=school_size_label)

grouped_size = joined_schol_data.groupby("binned")

avg_math_score = grouped_size['math_score'].mean()

avg_reading_score = grouped_size['reading_score'].mean()

math_pass_pct = joined_schol_data[joined_schol_data['math_score'] >= 70].groupby("binned")["Student ID"].count()*100/grouped_size["Student ID"].count()

reading_pass_pct = joined_schol_data[joined_schol_data['reading_score'] >= 70].groupby("binned")["Student ID"].count()*100/grouped_size["Student ID"].count()

math_read_pass_pct = joined_schol_data[(joined_schol_data['reading_score'] >= 70) & (joined_schol_data['math_score'] >= 70)].groupby("binned")["Student ID"].count()*100/grouped_size["Student ID"].count()

scores_by_school_size = pd.DataFrame({
    "Average Math Score": avg_math_score.round(2),
    "Average Reading Score": avg_reading_score.round(2),
    "% Passing Math": math_pass_pct.round(2),
    "% Passing Reading": reading_pass_pct.round(2),
    "% Overall Passing": math_read_pass_pct.round(2)
})

scores_by_school_size
```
